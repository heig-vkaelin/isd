```python
def is_prime(n):
    """Check if a given number is prime."""
    return n > 1 and all(n % i for i in range(2, int(math.sqrt(n)) + 1))
# Vectorize the function to apply it on a numpy array.
# This will return a mask that we can apply on the original array.
is_prime_np = np.vectorize(is_prime)
```

```python
x = np.arange(101)
```

```python
x[is_prime_np(x)]
```

## Exercise IV - Calcul d'un histogramme en Python

```python
from collections import Counter
from typing import Tuple


def compute_frequency(data: np.ndarray) -> Tuple[np.ndarray, np.ndarray]:
    """
    Compute the frequency of each item.
    
    Parameters
    ----------
    data : np.ndarray
        Original data to compute the frequency from.
        
    Returns
    -------
    Tuple of two numpy arrays
        Tuple with two numpy arrays. The first numpy array contains the unique values 
        and the second one their count.
    """
    return np.unique(data, return_counts=True)


def compute_hist(data: np.ndarray, bins: int = 10) -> Tuple[np.ndarray, np.ndarray]:
    """
    Compute the histogram.
    
    Parameters
    ----------
    data : np.ndarray
        Original data to compute the histogram from.
    bins : int, default 10
        Number of bins to use in the histogram.
    
    Returns
    -------
    Tuple of two numpy arrays
        Tuple with two numpy arrays. The first numpy array contains the number of element for each bins
        and the second one the limit of the bins.
    """
    values, freqs = compute_frequency(data)
    bins_edges = np.linspace(values.min(), values.max(), bins + 1)
    counts = np.zeros(bins + 1)
        
    # Look in which bin each value is.
    for v, f in zip(values, freqs):
        counts[np.searchsorted(bins_edges, v)] += f
        
    return counts, bins_edges
```

```python
N_BINS = 10
data = np.array([1, 1, 1, 2, 3, 3, 4, 5, 5, 7, 7, 7, 7, 7, 10])
```

```python
hist, edges = compute_hist(data, bins=N_BINS)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(edges, hist)
plt.title('Bar plot with our histogram');
```

```python
plt.figure(figsize=(12, 6))
plt.hist(data, bins=N_BINS)
plt.title('Histogram with the origin data');
```
